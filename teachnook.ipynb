# Visualization

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

#### Distribution of Age: Histogram and KDE plot

```python
# Set the aesthetic style of the plots
sns.set(style="whitegrid")

# Distribution of Age
plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
sns.histplot(data['Age'], kde=True, bins=30, color='skyblue')
plt.title('Age Distribution')

plt.subplot(1, 2, 2)
sns.histplot(data[data['Response'] == 1]['Age'], kde=True, bins=30, color='green', label='Responded')
sns.histplot(data[data['Response'] == 0]['Age'], kde=True, bins=30, color='red', label='Not Responded')
plt.legend()
plt.title('Age Distribution by Response')

plt.show()
```

#### Gender Distribution: Count plot.

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Gender', hue='Response', data=data, palette='Set1')
plt.title('Gender Distribution by Response')
plt.show()
```

#### Vehicle Age and Vehicle Damage: Count plots.

```python
plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
sns.countplot(x='Vehicle_Age', hue='Response', data=data, palette='Set1')
plt.title('Vehicle Age Distribution by Response')

plt.subplot(1, 2, 2)
sns.countplot(x='Vehicle_Damage', hue='Response', data=data, palette='Set1')
plt.title('Vehicle Damage Distribution by Response')
```

#### Annual Premium: Histogram and KDE plot.

```python
plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
sns.histplot(data['Annual_Premium'], kde=True, bins=30, color='blue')
plt.title('Annual Premium Distribution')

plt.subplot(1, 2, 2)
sns.histplot(data[data['Response'] == 1]['Annual_Premium'], kde=True, bins=30, color='green', label='Responded')
sns.histplot(data[data['Response'] == 0]['Annual_Premium'], kde=True, bins=30, color='red', label='Not Responded')
plt.legend()
plt.title('Annual Premium Distribution by Response')

plt.show()
```

#### Correlation Matrix: Heatmap to show correlations between numerical features.

```python
plt.figure(figsize=(12, 10))
# Convert categorical variables to numerical
data_encoded = pd.get_dummies(data, columns=['Gender', 'Vehicle_Age', 'Vehicle_Damage'], drop_first=True)
sns.heatmap(data_encoded.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

# Modeling

```python
data = pd.get_dummies(data, columns=['Gender', 'Vehicle_Age', 'Vehicle_Damage'], drop_first=True)
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

# Split the data into features and target variable
X = data.drop(['id', 'Response'], axis=1)
y = data['Response']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train a Logistic Regression model
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)

print(f'Accuracy: {accuracy}')
print(f'Classification Report:\n{report}')
```

#### Accuracy: The percentage of correctly predicted responses out of all predictions.Classification Report: Detailed metrics like precision, recall, and F1-score for both the positive and negative classes.
